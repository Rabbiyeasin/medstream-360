```python
# =============================================================================
# DAY 3: Bronze Layer Verification (WSL Optimized)
# =============================================================================
from pyspark.sql import SparkSession
from delta.tables import *

# Initialize Spark with Delta Lake support
spark = SparkSession.builder \
    .appName("MedStream360 - Bronze Verification") \
    .config("spark.sql.extensions", "io.delta.sql.DeltaSparkSessionExtension") \
    .config("spark.sql.catalog.spark_catalog", "org.apache.spark.sql.delta.catalog.DeltaCatalog") \
    .getOrCreate()

# Path to Bronze layer (relative from notebooks folder)
bronze_path = "../data/bronze/vitals"

try:
    df_bronze = spark.read.format("delta").load(bronze_path)
    
    print("✅ SUCCESS: Bronze Layer Loaded from Delta Table!")
    print(f"Total Records: {df_bronze.count()}")
    print("\nSample Data:")
    df_bronze.show(5)
    
    print("\nSchema:")
    df_bronze.printSchema()
    
except Exception as e:
    print(f"❌ Error loading Bronze layer: {e}")
    print("\nPossible reasons:")
    print("1. Kafka consumer hasn't run yet → Run the producer + consumer first")
    print("2. Path is wrong → Check if data/bronze/vitals exists")
    print("3. No data written → Run consumer for 30-60 seconds")
```
